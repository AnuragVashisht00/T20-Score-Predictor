```python
import pickle
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings('ignore')
```

```python
df=pickle.load(open("dataset_level2.pkl","rb"))
```

```python
df
```

```python
df["city"].isnull().sum()
```

```python
cities=np.where(df["city"].isnull(),df["venue"].str.split().apply(lambda x:x[0]),df["city"])
```

```python
df["city"]=cities
```

```python
df.isnull().sum()
```

```python
df.drop(columns=["venue"],inplace=True)
```

```python
eligible_cities=df["city"].value_counts()[df["city"].value_counts()>600].index.tolist()
```

```python
df=df[df["city"].isin(eligible_cities)]
```

```python
df
```

```python
df["current_score"]=df.groupby("match_id").cumsum()["runs"]
```

```python
df
```

```python
df["over"]=df["ball_of_match"].apply(lambda x:str(x).split(".")[0])
df["ball"]=df["ball_of_match"].apply(lambda x:str(x).split(".")[1])
```

```python
df["ball_bowled"]=(df["over"].astype("int")*6)+df["ball"].astype("int")
```

```python
df
```

```python
df["ball_left"]=120 - df["ball_bowled"]
df["ball_left"]=df["ball_left"].apply(lambda x:0 if x<0 else x)
```

```python
df
```

```python
df['player_dismissed'] = df['player_dismissed'].apply(lambda x:0 if x=='0' else 1)
df['player_dismissed'] = df['player_dismissed'].astype('int')
df['player_dismissed'] = df.groupby('match_id').cumsum()['player_dismissed']
df['wickets_left'] = 10 - df['player_dismissed']
```

```python
df
```

```python
df["crr"]=(df["current_score"]*6)/df["ball_bowled"]
```

```python
df
```

```python
groups = df.groupby('match_id')

match_ids = df['match_id'].unique()
last_five = []
for id in match_ids:
    last_five.extend(groups.get_group(id).rolling(window=30).sum()['runs'].values.tolist())
```

```python
df["last_five"]=last_five
```

```python
df
```

```python
final_df=df.groupby("match_id").sum()["runs"].reset_index().merge(df,on="match_id")
```

```python
final_df=final_df[["batting_team","bowling_team","city","current_score","ball_left","wickets_left","crr","last_five","runs_x"]]
```

```python
final_df.isnull().sum()
```

```python
final_df.dropna(inplace=True)
```

```python
final_df
```

```python
final_df=final_df.sample(final_df.shape[0])
```

```python
final_df
```

```python
x=final_df.drop(columns=["runs_x"])
y=final_df["runs_x"]
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=1)
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import RandomForestRegressor
from xgboost import XGBRegressor
from sklearn.metrics import r2_score,mean_absolute_error
```

```python
trf = ColumnTransformer([('trf',OneHotEncoder(sparse=False,drop='first'),['batting_team','bowling_team','city'])],remainder='passthrough')
```

```python
pipe = Pipeline(steps=[
    ('step1',trf),
    ('step2',StandardScaler()),
    ('step3',XGBRegressor(n_estimators=1000,learning_rate=0.2,max_depth=12,random_state=1))
])
```

```python
pipe.fit(x_train,y_train)
y_pred = pipe.predict(x_test)
print(r2_score(y_test,y_pred))
print(mean_absolute_error(y_test,y_pred))
```

```python
pickle.dump(pipe,open('pipe.pkl','wb'))
```
